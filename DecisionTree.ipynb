```
from sklearn import tree
import pandas as pd
```

Data part.

```
data={
      'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny'],
    'Temperature': ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild'],
    'Humidity': ['High','High','High','High','Normal','Normal','Normal','High'],
    'Wind': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak'],
    'PlayTennis': ['No','No','Yes','Yes','Yes','No','Yes','No']
}
```

```
df=pd.DataFrame(data)
```

simple preprocessing part. label encoder convert caategorical text lables into numerical values.


```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df.columns:
  df[col]=le.fit_transform(df[col])
```

```
x=df.drop('PlayTennis', axis =1)
y=df['PlayTennis']
```

```
lef=tree.DecisionTreeClassifier(criterion='entropy')
lef=lef.fit(x,y)
```

```
tree.plot_tree(lef, feature_names=x.columns, class_names=['No','Yes'], filled=True)
```
